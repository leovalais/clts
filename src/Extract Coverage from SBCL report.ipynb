```python
import pandas as pd
```

```python
covhtml = "coverage/cover-index.html"
```

```python
df = pd.read_html(covhtml)[0]
```

```python
df
```

```python
df2 = df.drop([0, 1]).drop(columns=[0])
```

```python
for i in range(1, len(df2.columns) + 1):
    df2[i] = df2[i].str.extract(r'(\d+)', expand=True).dropna().astype(int)
```

```python
dff = df2.dropna()
dff.columns = "cov_exp cov_tot cov_per banch_cov branch_tot branch_per".split(' ')
```

```python
dff
```

```python
def mean_cov(axis, remove_epsilon=3):
    return dff[dff[axis] > remove_epsilon][axis].mean()
```
